```python
def get_score(server,name):
    search = "https://raider.io/api/v1/characters/profile?region=eu&realm="+server+"&name="+name+"&fields=mythic_plus_scores_by_season%3Acurrent"
    response = requests.get(search)
    if response.status_code == 200:
        return response.json()
    else:
        return
```

```python
cols = ['id','name','achievement_points','equipped_item_level','gender.name','faction.name','race.name','active_spec.name','realm.slug','pvp_summary.href','statistics.href','mythic_keystone_profile.href','equipment.href','active_title.name','professions.href','covenant_progress.chosen_covenant.name','covenant_progress.renown_level','covenant_progress.soulbinds.href']

def character_details(realm,character):
    search = "https://eu.api.blizzard.com/profile/wow/character/"+realm+"/"+character+"?namespace=profile-eu&locale=en_GB&access_token="+token
    response = requests.get(search)
    return pd.json_normalize(response.json())[cols]
```

```python
def get_gear(realm,character):
    search = "https://eu.api.blizzard.com/profile/wow/character/"+realm+"/"+character+"/equipment?namespace=profile-eu&locale=en_GB&access_token="+token
    response = requests.get(search)
    if response.status_code == 200:
        return response.json()
    else:
        return 404
```

```python
# Talents
def get_talents(realm,character):
    search = "https://eu.api.blizzard.com/profile/wow/character/"+realm+"/"+character.lower()+"/specializations?namespace=profile-eu&locale=en_GB&access_token="+token
    response = requests.get(search)
    if response.status_code == 200:
        test = pd.json_normalize(response.json(),record_path = ['specializations'])
        test = test[ test['specialization.name'] == 'Protection']
        test1 = pd.json_normalize(test['talents'].iloc[0])['talent.name']
        test2 = pd.json_normalize(test['pvp_talent_slots'].iloc[0])['selected.talent.name']
        test3 = list(test1.values) + list(test2.values)
        return pd.DataFrame(test3).T
    else:
        return
```

```python
def get_conduits(realm,character):
    search = "https://eu.api.blizzard.com/profile/wow/character/hellfire/skyl%C3%ADne/soulbinds?namespace=profile-eu&locale=en_GB&access_token="+token
    response = requests.get(search)
    if response.status_code == 200:
        return response.json()
    else:
        return 404
```

```python
def get_pvp(realm,char):
    search = 'https://eu.api.blizzard.com/profile/wow/character/'+realm+'/'+char+'/pvp-summary?namespace=profile-eu&locale=en_GB&access_token='+token
    response = requests.get(search)
    data = pd.json_normalize(response.json())
    set1 = data[['character.id','character.name','character.realm.slug','honor_level','honorable_kills']]
    
    set2 = pd.json_normalize( data['pvp_map_statistics'][0] )
    set2['win_rate'] = round( (set2['match_statistics.won'] / set2['match_statistics.played']) * 100 )
    
    set3 = set2[['world_map.name','win_rate']]
    set3 = set3.set_index('world_map.name').T
    set3 = set3.add_suffix(' win_rate')
    set3['character.id'] = set1['character.id'][0]
    set3['character.name'] = set1['character.name'][0]
    set3['character.realm.slug'] = set1['character.realm.slug'][0]
    
    set4 = set2[['world_map.name','match_statistics.played']]
    set4 = set4.set_index('world_map.name').T
    set4 = set4.add_suffix(' played')
    set4['character.id'] = set1['character.id'][0]
    set4['character.name'] = set1['character.name'][0]
    set4['character.realm.slug'] = set1['character.realm.slug'][0]
    
    set5 = set2[['world_map.name','match_statistics.won']]
    set5 = set5.set_index('world_map.name').T
    set5 = set5.add_suffix(' won')
    set5['character.id'] = set1['character.id'][0]
    set5['character.name'] = set1['character.name'][0]
    set5['character.realm.slug'] = set1['character.realm.slug'][0]
    
    pvp = pd.concat([set1,set3,set4,set5])
    
    pvp = pvp.groupby('character.id').max().reset_index()
    return pvp
```

```python
def get_score(server,name):
    search = "https://raider.io/api/v1/characters/profile?region=eu&realm="+server+"&name="+name+"&fields=mythic_plus_scores_by_season%3Acurrent"
    response = requests.get(search)
    if response.status_code == 200:
        score_response = pd.json_normalize( response.json(), record_path = 'mythic_plus_scores_by_season' )['scores.all'].values
        return score_response[0]
    else:
        return
```

```python
realms = list(pd.read_csv('parent_realms.csv')['id'].unique())
for realm in realms:
    print("Realm - - - "+str(realm))
    k = 1
    pala_boards = pd.read_csv("C:\\Users\\patri\\Desktop\\Raider io\\realms\\"+str(realm)+".csv")
    gear_df = pd.DataFrame()
    talents_df = pd.DataFrame()
    details_df = pd.DataFrame()
    conduits_df = pd.DataFrame()
    pvp_df = pd.DataFrame()
    for i, j in pala_boards.iterrows():
        char_id = j['profile.id']
        char_name = j['profile.name']
        char_server = j['profile.realm.slug']
        print(str(k) + "/"+str(pala_boards['profile.name'].size))
        k+=1
        try:
            # Get Character Details
            temp=character_details(char_server,char_name.lower())
            active_spec = str(temp['active_spec.name'][0])
            if active_spec == 'Protection' : # Get current Gear
                # Character Gear
                char_id = pd.json_normalize(get_gear(char_server,char_name.lower()))['character.id']
                temp = pd.json_normalize(get_gear(char_server,char_name.lower()) ,record_path=['equipped_items'])
                temp = temp[['name','slot.name']].T
                temp.columns = temp.iloc[1]
                temp = temp.drop(temp.index[1]).reset_index(drop=True)
                temp['character.id'] = char_id
                gear_df = gear_df.append(pd.DataFrame(temp))
                # Get Talents
                temp = get_talents(char_server,char_name)
                temp['character.id'] = char_id
                temp.columns=['L15','L25','L30','L35','L40','L45','L50','pvp_1','pvp_2','pvp_3','character.id']
                talents_df = talents_df.append(temp, ignore_index=True)
                # Get Details
                temp = character_details(char_server,char_name.lower())
                temp = temp.rename(columns={'id':'character.id'})
                details_df = details_df.append(pd.DataFrame(temp),ignore_index=True)
                # Get Soulbinds
                temp = pd.json_normalize(get_conduits(char_server,char_name.lower()) ,record_path=['soulbinds'])
                temp = temp[temp['is_active']==True]
                soulbinds = list(temp['soulbind.name'])
                temp = pd.json_normalize(temp['traits'][0])
                traits = list(temp[ ~temp['trait.name'].isna() ]['trait.name'])
                things = list(temp[ temp['trait.name'].isna() ]['conduit_socket.socket.conduit.name'])
                conduits = pd.DataFrame(data=[soulbinds + traits + things])
                conduits['character.id'] = char_id
                conduits_df = conduits_df.append(conduits,ignore_index=True)
                # pvp
                temp = get_pvp(char_server,char_name.lower())
                temp['character.id'] = char_id
                pvp_df = pvp_df.append(temp,ignore_index=True)
        except KeyError:
            pass
        except TypeError:
            pass
        except ValueError:
            pass
    # Export realms info
    gear_df.to_csv('gear\\'+str(realm)+'.csv',index=False)
    talents_df.to_csv('talents\\'+str(realm)+'.csv',index=False)
    details_df.to_csv('details\\'+str(realm)+'.csv',index=False)
    conduits_df.to_csv('conduits\\'+str(realm)+'.csv',index=False)
    pvp_df.to_csv('pvp\\'+str(realm)+'.csv',index=False)
```

```python
realms = list(pd.read_csv('parent_realms.csv')['id'].unique())
for realm in realms:
    print("Realm - - - "+str(realm))
    k = 1
    pala_boards = pd.read_csv("C:\\Users\\patri\\Desktop\\Raider io\\realms\\"+str(realm)+".csv")
    scores_df = pd.DataFrame()
    for i, j in pala_boards.iterrows():
        char_name = j['profile.name']
        char_server = j['profile.realm.slug']
        char_id = j['profile.id']
        print(str(k) + "/"+str(pala_boards['profile.name'].size))
        k+=1
        try:
            # Get Scores
            char_score = get_score(char_server,char_name)
            temp = pd.DataFrame(columns=['character.id','profile.name','profile.slug','score'],
                               data=[[char_id,char_name,char_server, char_score]])
            
            scores_df = scores_df.append(temp)
        except KeyError:
            pass
        except TypeError:
            pass
    # Export realms info
    scores_df.to_csv('scores\\'+str(realm)+'.csv',index=False)
```
